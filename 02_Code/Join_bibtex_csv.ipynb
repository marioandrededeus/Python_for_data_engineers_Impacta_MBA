# Imports

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 300)
```

# Load CSV merged files

```python
df_merged_csv = pd.read_csv('../01_Datasets/csv/df_merged.csv', sep=';', low_memory=False)
df_merged_csv.head()
```

```python
df_merged_csv.issn.head()
```

## Split issn column

```python
df_merged_csv['issn'] = df_merged_csv['issn'].map(lambda x: x.split(', ') if x==x else np.nan)
df_merged_csv['issn1'] = df_merged_csv['issn'].map(lambda x: x[0] if x == x else np.nan)
df_merged_csv['issn2'] = df_merged_csv['issn'].map(lambda x: np.nan if x !=x else x[1] if len(x) > 1 else x[0])
df_merged_csv['issn3'] = df_merged_csv['issn'].map(lambda x: np.nan if x !=x else x[2] if len(x) > 2 else x[0])
df_merged_csv.drop('issn', axis = 1, inplace = True)
df_merged_csv[['issn1', 'issn2','issn3']]
```

# Load bibtex file

```python
df_bibtex = pd.read_csv('../03_OutputFiles/df_all.csv',sep=';')
df_bibtex.head()
```

```python
df_bibtex['issn'] = df_bibtex['issn'].map(lambda x: x.replace('-','') if x == x else np.nan)
df_bibtex['issn']
```

## Title formatting

```python
df_bibtex['title'] = df_bibtex['title'].map(lambda x: x.capitalize() if x == x else np.nan)
df_bibtex.head()
```

# Merging CSV and bibtex files

```python
print(df_merged_csv.shape, df_bibtex.shape)
```

```python
#On "issn1"
df_merged_csv_bib_issn1 = df_merged_csv.merge(df_bibtex, how = 'outer', left_on='issn1', right_on = 'issn', copy=False, indicator = True)
df_merged_csv_bib_issn1 = df_merged_csv_bib_issn1.loc[(df_merged_csv_bib_issn1._merge == 'both') & (df_merged_csv_bib_issn1.issn.notna())]
df_merged_csv_bib_issn1.drop('_merge', axis = 1, inplace = True)
print('On issn1: \n', df_merged_csv_bib_issn1.shape, '\n')

#On "issn2"
df_merged_csv_bib_issn2 = df_merged_csv.merge(df_bibtex, how = 'outer', left_on='issn2', right_on = 'issn', copy=False, indicator = True)
df_merged_csv_bib_issn2 = df_merged_csv_bib_issn2.loc[(df_merged_csv_bib_issn2._merge == 'both') & (df_merged_csv_bib_issn2.issn.notna())]
df_merged_csv_bib_issn2.drop('_merge', axis = 1, inplace = True)
print('On issn2: \n', df_merged_csv_bib_issn2.shape, '\n')

#On "issn3"
df_merged_csv_bib_issn3 = df_merged_csv.merge(df_bibtex, how = 'outer', left_on='issn3', right_on = 'issn', copy=False, indicator = True)
df_merged_csv_bib_issn3 = df_merged_csv_bib_issn3.loc[(df_merged_csv_bib_issn3._merge == 'both') & (df_merged_csv_bib_issn3.issn.notna())]
df_merged_csv_bib_issn3.drop('_merge', axis = 1, inplace = True)
print('On issn3: \n', df_merged_csv_bib_issn3.shape, '\n')
```

```python
df_merged_csv_bib_issn1_2_3 = pd.concat([df_merged_csv_bib_issn1,df_merged_csv_bib_issn2,df_merged_csv_bib_issn3])
df_merged_csv_bib_issn1_2_3.sort_values('issn', inplace = True)
cols = np.sort(df_merged_csv_bib_issn1_2_3.columns.tolist())
df_merged_csv_bib_issn1_2_3 = df_merged_csv_bib_issn1_2_3[cols]

print('Before deduplication: ', df_merged_csv_bib_issn1_2_3.shape)
df_merged_csv_bib_issn1_2_3_deduplicate = df_merged_csv_bib_issn1_2_3.drop_duplicates(subset = 'issn')
df_merged_csv_bib_issn1_2_3_deduplicate.reset_index(drop = True, inplace = True)
print('After deduplication: ', df_merged_csv_bib_issn1_2_3_deduplicate.shape)

df_merged_csv_bib_issn1_2_3_deduplicate
```

# Export merged files

```python
df_merged_csv_bib_issn1_2_3_deduplicate.to_csv('../01_Datasets/merged/df_merged_csv_bib.csv', sep=';', index = False)
```
