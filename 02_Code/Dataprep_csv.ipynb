# Imports

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 300)
```

# Load jcs.csv

```python
path_jcs = '/Users/amigosdadancamooca/Documents/Impacta/Python_for_Data_Engineers/01_Datasets/csv/jcs_2020.csv'
df_jcs = pd.read_csv(path_jcs, sep=';', low_memory = False, thousands = ',', decimal='.')
print(df_jcs.shape)
df_jcs.head()
```

## Dropna

```python
df_jcs.isna().sum()
```

```python
df_jcs.dropna(axis = 1, inplace = True) 
df_jcs.isna().sum()
```

## Columns renaming

```python
cols_jcs = df_jcs.columns.str.lower()
cols_jcs = list(map(lambda x: x.replace(' ', '_'), cols_jcs))
df_jcs.columns = cols_jcs
df_jcs.head()
```

## Title formatting

```python
df_jcs.rename(columns={'full_journal_title' : 'title'}, inplace = True)
df_jcs['title'] = df_jcs['title'].map(lambda x: x.capitalize())
df_jcs
```

# Load scimagojr 2020.csv

```python
path_scimagojr = '/Users/amigosdadancamooca/Documents/Impacta/Python_for_Data_Engineers/01_Datasets/csv/scimagojr_2020.csv'
df_scimagojr = pd.read_csv(path_scimagojr, sep=';', low_memory = False, thousands = ',')
print(df_scimagojr.shape)
df_scimagojr.head(2)
```

```python
df_scimagojr.isna().sum()
```

## Columns Renaming

```python
cols_scimagojr = df_scimagojr.columns.str.lower()
cols_scimagojr = list(map(lambda x: x.replace(' ', '_'), cols_scimagojr))
df_scimagojr.columns = cols_scimagojr
df_scimagojr.head()
```

## Title formatting

```python
df_scimagojr['title'] = df_scimagojr['title'].map(lambda x: x.capitalize())
df_scimagojr
```

# Merging CSV files

```python
print(df_jcs.shape, df_scimagojr.shape)
```

```python
df_merged_csv = df_jcs.merge(df_scimagojr, how = 'outer', on='title',copy=False, indicator = True)
print(df_merged_csv.shape)
print(df_merged_csv._merge.value_counts())
df_merged_csv.head()
```

```python
#As colunas "rank" possuem numeros diferentes em cada base para um mesmo "title", e por isso serão dropadas
#Também será dropada a coluna "_merge" utilizada para verificar o numero de coincidencias entre as bases
df_merged_csv.drop(['rank_x', 'rank_y', '_merge'], axis = 1, inplace = True, errors = 'ignore')
df_merged_csv.columns.tolist()
```

# Export CSV merged files

```python
df_merged_csv.to_csv('../01_Datasets/csv/df_merged.csv', sep=';', index = False)
```
